```python
# импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv') 
credits.columns = ['id','tittle','cast','crew']
movies = movies.merge(credits,on='id')
```

```python
# столбчатую диаграмма распределения иностранных фильмов (не англ.)
ol = movies.groupby(['original_language']).count().sort_values(by='id', ascending=False).reset_index()
ol = ol[ol['original_language'] != 'en'].head(10)
plt.barh(ol['original_language'], ol['id'])
plt.grid()
```

```python
# круговая диаграмма распределения языков для фильмов (top-4)
mm = movies['original_language'].value_counts().head(5)
mm.plot(kind='pie', y='id')
```
