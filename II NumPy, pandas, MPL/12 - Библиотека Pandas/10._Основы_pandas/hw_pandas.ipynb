Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')

# ratings.info()

#количество уникальных фильмов
# ratings.movieId.nunique()

# Фильмы, которым кто-то поставил 5
films_rating_5 = ratings[ratings.rating == 5]

#Группировка фильмов (movieId) по количеству (count_films_rating_5) 
films_rating_5_grouped = films_rating_5.groupby('movieId').count()

# сортировка по убыванию, лучшие фильмы вверху
films_rating_5_grouped_sorted = films_rating_5_grouped.sort_values('rating', ascending=False)

# слияние данных по movieId для получения названий фильмов
best_movies = films_rating_5_grouped_sorted.merge(movies, on='movieId', how='left')

# волнительный момент, лучший фильм - Побег из Шоушенка (хотя мне больше Форрест Гамп зашел)
best_movies.head(1)
best_movies.head(5)
```

Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.


```python
power_data = pd.read_csv('power.csv')
list_category = [2, 12, 21]

# фильтр по ТЗ
power_data_filtered = power_data[((power_data['year'] >= 2005) & (power_data['year'] <= 2010)) & \
                                 ((power_data['category'] == 2) | \
                                   (power_data['category'] == 12) | \
                                   (power_data['category'] == 21)) & \
                                 ((power_data['country'] == 'Latvia') | \
                                  (power_data['country'] == 'Lithuania') | \
                                  (power_data['country'] == 'Estonia')) & \
                                 (power_data['quantity'] > 0)
                                ]
# 
print(power_data_filtered.groupby('country')['quantity'].sum())
# Суммарное потребление стран          
print(f"Суммарное потребление: {power_data_filtered['quantity'].sum()}")
```
