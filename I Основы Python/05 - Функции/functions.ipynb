Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
def search_peaple(docs, number):
    for doc in docs:
        if doc['number'] == number:
            return doc['name']

def search_shelf(direct, number):
    for key, shelf_list in direct.items():
        if number in shelf_list:
            return key

def list_doc(docs):
    print('Список документов:')
    for doc in docs:
        print(f'{doc["type"]} "{doc["number"]}" "{doc["name"]}"')

def list_shelf(direct):
    return direct
  

def add_shelf(direct, number):
    if number not in direct:
        direct[number] = []
        return True
    else:
        return False

def delete_doc(docs, direct, number):
    state = False
    for doc in docs:
        if doc['number'] == number:
            docs.remove(doc)
            state = True
    for index, list_doc in direct.items():
        if number in list_doc:
            direct[index].remove(number)
            state = True
    return state
    
def move_doc(direct, number, new_shelf):
    if new_shelf not in direct:
        return False
    for index, list_doc in direct.items():
        if number in list_doc:
            direct[index].remove(number)
            direct[new_shelf].append(number)
            return True
            
def add_doc(docs, direct, type_doc, number, name, shelf):
#     list_type = ('passport', 'invoice', 'insurance')
#     if type_doc not in list_type:
#         return False
    if shelf not in direct:
        return False
    doc = {'type': type_doc, 'number': number, 'name': name}
    docs.append(doc)
    direct[shelf].append(number)
    return True


def help_print():
    print(    
        '''
        Cправка по командам:
        p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;
        s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится
        l – list – команда, которая выведет список всех документов в формате passport "2207 876234" "Василий Гупкин"
        as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень.
        d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок.
        m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую.
        a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.
        ls - list shelf - команда выведет список полок и документов на них
        '''
    )    
    
def main():
    while True:
        user_input = input('Введите команду: (h - help, q - quit) >').lower()
        if user_input in ('q', 'quit', 'exit'):
            break
        elif user_input in ('l', 'list'):
            list_doc(documents)
        elif user_input in ('p', 'peaple'):
            peaple = search_peaple(documents, str(input('Поиск ФИО по номеру. Введите номер:')))
            print(peaple if peaple != None else 'Нет данных')
        elif user_input in ('s', 'shelf'):
            shelf = search_shelf(directories, str(input('Поиск полки по номеру. Введите номер:')))            
            print(shelf if shelf != None else 'Нет данных')
        elif user_input in ('ls', 'list shelf'):    
            print(list_shelf(directories))
        elif user_input in ('as', 'add shelf'):
            shelf = str(input('Добавим новую полку. Введите номер полки:'))
            print((f'Полка №{shelf} добавлена') if add_shelf(directories, shelf) else 'Такая полка уже существует')
        elif user_input in ('d', 'delete'):
            doc = str(input('Удаление документа. Введите номер документа:'))
            print('Документ удален' if delete_doc(documents, directories, doc) else 'Документ не найден')
        elif user_input in ('m', 'move'):            
            doc = str(input('Перемещение документа. Введите номер документа:'))
            shelf = str(input('Перемещение документа. Введите номер полки:'))
            print('Документ перемещен' if move_doc(directories, doc, shelf) else 'Ошибка ввода')
        elif user_input in ('a', 'add'):
            type_doc = str(input('Создание документа. Введите тип документа:'))
            number_doc = str(input('Создание документа. Введите номер документа:'))
            name_doc = str(input('Создание документа. Введите имя владельца:'))
            shelf = str(input('Создание документа. Введите номер полки:'))
            print('Документ создан' if add_doc(documents, directories, type_doc, number_doc, name_doc, shelf) else 'Нет такой полки')            
        else:
            print('Команда не найдена')
            help_print()
            
main()
```
